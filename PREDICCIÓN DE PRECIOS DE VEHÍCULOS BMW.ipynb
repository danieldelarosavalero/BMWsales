### 2.2 Formulaci√≥n como Problema de Machine Learning (P2)

#### Caracter√≠sticas del Problema

**Tipo de aprendizaje:** Aprendizaje Supervisado

**Tipo de tarea:** Regresi√≥n (predicci√≥n de valor continuo)

**Formalizaci√≥n matem√°tica:**

Tenemos un dataset $D = \{(x_i, y_i)\}_{i=1}^{n}$ donde:
- $x_i \in \mathbb{R}^d$: vector de d features del veh√≠culo i
- $y_i \in \mathbb{R}^+$: precio del veh√≠culo i (variable continua positiva)
- n = 10,781 (n√∫mero de ejemplos)
- d = 8 (n√∫mero de features despu√©s de preprocesamiento)

**Objetivo:** Aprender una funci√≥n $f: \mathbb{R}^d \rightarrow \mathbb{R}^+$ tal que:

$$f(x) \approx y$$

Minimizando una funci√≥n de p√©rdida (loss function), t√≠picamente MSE:

$$L = \frac{1}{n}\sum_{i=1}^{n}(y_i - f(x_i))^2$$

#### Variables del Problema

**Variable objetivo (Target):**

| Variable | Tipo | Descripci√≥n | Rango | Estad√≠sticas |
|----------|------|-------------|-------|--------------|
| **price** | Continua | Precio del veh√≠culo en ¬£ | ¬£1,200 - ¬£123,456 | Media: ¬£22,733<br>Mediana: ¬£20,462<br>Std: ¬£11,415 |

**Features (Variables predictoras):**

| # | Variable | Tipo | Descripci√≥n | Valores √∫nicos | Ejemplos |
|---|----------|------|-------------|----------------|----------|
| 1 | **model** | Categ√≥rica | Modelo BMW | 24 | "3 Series", "5 Series", "X3", "M3" |
| 2 | **year** | Num√©rica | A√±o de fabricaci√≥n | 25 | 1996-2020 |
| 3 | **transmission** | Categ√≥rica | Tipo de transmisi√≥n | 3 | "Manual", "Automatic", "Semi-Auto" |
| 4 | **mileage** | Num√©rica | Kilometraje | 8,086 | 0 - 323,000 km |
| 5 | **fuelType** | Categ√≥rica | Tipo de combustible | 5 | "Diesel", "Petrol", "Hybrid", "Electric", "Other" |
| 6 | **tax** | Num√©rica | Impuesto anual | 38 | ¬£0 - ¬£600 |
| 7 | **mpg** | Num√©rica | Consumo (millas/gal√≥n) | 102 | 5.5 - 470.8 mpg |
| 8 | **engineSize** | Num√©rica | Tama√±o motor en litros | 17 | 0.0 - 6.6 L |

**Nota sobre 'mpg':** Valores muy altos (>100) t√≠picamente corresponden a veh√≠culos h√≠bridos/el√©ctricos.

#### M√©tricas de Evaluaci√≥n

Para evaluar nuestros modelos de regresi√≥n, utilizaremos las siguientes m√©tricas est√°ndar:

**1. MAE (Mean Absolute Error) - Error Absoluto Medio:**

$$MAE = \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y}_i|$$

- **Interpretaci√≥n:** Error promedio en ¬£ (libras esterlinas)
- **Ventaja:** F√°cil de interpretar, misma unidad que el target
- **Desventaja:** No penaliza errores grandes desproporcionadamente
- **Objetivo:** < ¬£2,000 (< 10% del precio medio)

**2. RMSE (Root Mean Squared Error) - Ra√≠z del Error Cuadr√°tico Medio:**

$$RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}$$

- **Interpretaci√≥n:** Error cuadr√°tico medio (penaliza errores grandes)
- **Ventaja:** Penaliza outliers m√°s que MAE
- **Desventaja:** M√°s sensible a valores at√≠picos
- **Objetivo:** < ¬£2,500

**3. R¬≤ Score (Coeficiente de Determinaci√≥n):**

$$R^2 = 1 - \frac{\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}$$

- **Interpretaci√≥n:** Proporci√≥n de varianza explicada por el modelo
- **Rango:** 0 (modelo in√∫til) a 1 (predicci√≥n perfecta)
- **Ventaja:** Normalizado, comparable entre datasets
- **Objetivo:** > 0.85 (85% de varianza explicada)

**4. MAPE (Mean Absolute Percentage Error) - Error Porcentual Absoluto Medio:**

$$MAPE = \frac{100\%}{n}\sum_{i=1}^{n}\left|\frac{y_i - \hat{y}_i}{y_i}\right|$$

- **Interpretaci√≥n:** Error promedio en porcentaje
- **Ventaja:** Independiente de escala, f√°cil de comunicar
- **Desventaja:** Indefinido cuando $y_i = 0$, sesgado hacia underprediction
- **Objetivo:** < 8%

**M√©trica principal:** Usaremos **R¬≤ Score** como m√©trica principal para comparar modelos, complementado con RMSE para interpretar error absoluto.


#### Desaf√≠os Espec√≠ficos del Problema

**1. Variables categ√≥ricas con alta cardinalidad:**
- `model` tiene 24 categor√≠as √∫nicas
- Requiere encoding apropiado (One-Hot, Target Encoding, etc.)
- Riesgo de dimensionalidad excesiva con One-Hot Encoding

**2. Escalas muy diferentes entre features:**
- `year`: 1996-2020 (rango ~24)
- `mileage`: 0-323,000 (rango ~323,000)
- `price`: ¬£1,200-¬£123,456 (rango ~122,000)
- Requiere normalizaci√≥n/estandarizaci√≥n para algunos algoritmos (KNN, SVR, Neural Networks)

**3. Outliers en el target:**
- Veh√≠culos muy caros (>¬£80,000) son raros
- Pueden sesgar el entrenamiento
- Debemos decidir si eliminarlos, transformarlos (log), o usar algoritmos robustos

**4. Relaciones no lineales:**
- La depreciaci√≥n no es lineal con el tiempo
- El efecto del kilometraje tiene "umbrales" (ej: 100,000 km)
- Requiere algoritmos capaces de capturar no linealidad (√°rboles, redes neuronales)

**5. Interacciones entre variables:**
- El efecto del kilometraje depende del modelo (Serie 7 vs Serie 1)
- El efecto del tipo de combustible depende del a√±o (regulaciones cambiantes)
- Algunos algoritmos (RF, XGBoost) capturan esto autom√°ticamente

**6. Desequilibrio en distribuci√≥n de modelos:**
- Algunos modelos (Serie 3) son muy comunes
- Otros modelos (M8) son muy raros
- Puede causar que el modelo aprenda mejor para modelos comunes

**7. Features correlacionados:**
- `year` y `mileage` est√°n correlacionados (veh√≠culos antiguos ‚Üí m√°s km)
- `engineSize` y `fuelType` correlacionados (motores grandes ‚Üí gasolina)
- Puede causar multicolinealidad en regresi√≥n lineal

**Estrategias para abordar estos desaf√≠os:**
- Feature engineering cuidadoso
- Selecci√≥n de algoritmos apropiados
- Regularizaci√≥n (Ridge, Lasso)
- Ensembles (Random Forest, XGBoost)
- Validaci√≥n cruzada robusta

# AN√ÅLISIS Y VISUALIZACI√ìN DEL PRECIO


````python
print("=" * 80)
print("üí∞ AN√ÅLISIS DETALLADO DE LA VARIABLE OBJETIVO: PRICE")
print("=" * 80)

# Estad√≠sticas descriptivas
print("\nüìä Estad√≠sticas Descriptivas:")
print(f"   Media:           ¬£{df['price'].mean():,.2f}")
print(f"   Mediana:         ¬£{df['price'].median():,.2f}")
print(f"   Desv. Est√°ndar:  ¬£{df['price'].std():,.2f}")
print(f"   M√≠nimo:          ¬£{df['price'].min():,.2f}")
print(f"   M√°ximo:          ¬£{df['price'].max():,.2f}")
print(f"   Rango:           ¬£{df['price'].max() - df['price'].min():,.2f}")

# Cuartiles
print(f"\nüìà Cuartiles:")
print(f"   Q1 (25%):        ¬£{df['price'].quantile(0.25):,.2f}")
print(f"   Q2 (50%):        ¬£{df['price'].quantile(0.50):,.2f}")
print(f"   Q3 (75%):        ¬£{df['price'].quantile(0.75):,.2f}")

# IQR y outliers
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print(f"\nüéØ An√°lisis de Outliers (m√©todo IQR):")
print(f"   IQR:                ¬£{iqr:,.2f}")
print(f"   L√≠mite inferior:    ¬£{lower_bound:,.2f}")
print(f"   L√≠mite superior:    ¬£{upper_bound:,.2f}")
print(f"   Outliers detectados: {len(outliers)} ({len(outliers)/len(df)*100:.2f}%)")

# Asimetr√≠a y curtosis
from scipy import stats
skewness = stats.skew(df['price'])
kurtosis_val = stats.kurtosis(df['price'])

print(f"\nüìê Forma de la Distribuci√≥n:")
print(f"   Asimetr√≠a (Skewness): {skewness:.3f}")
if skewness > 0.5:
    print("      ‚Üí Distribuci√≥n sesgada a la DERECHA (cola larga hacia valores altos)")
elif skewness < -0.5:
    print("      ‚Üí Distribuci√≥n sesgada a la IZQUIERDA (cola larga hacia valores bajos)")
else:
    print("      ‚Üí Distribuci√≥n aproximadamente SIM√âTRICA")

print(f"   Curtosis:            {kurtosis_val:.3f}")
if kurtosis_val > 0:
    print("      ‚Üí Leptoc√∫rtica (pico m√°s agudo que distribuci√≥n normal)")
else:
    print("      ‚Üí Platic√∫rtica (pico m√°s plano que distribuci√≥n normal)")

print("=" * 80)

# Visualizaci√≥n completa
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
fig.suptitle('An√°lisis Completo de la Variable PRICE', fontsize=18, fontweight='bold')

# 1. Histograma
axes[0, 0].hist(df['price'], bins=50, edgecolor='black', alpha=0.7, color='steelblue')
axes[0, 0].axvline(df['price'].mean(), color='red', linestyle='--', linewidth=2, 
                   label=f'Media: ¬£{df["price"].mean():,.0f}')
axes[0, 0].axvline(df['price'].median(), color='green', linestyle='--', linewidth=2, 
                   label=f'Mediana: ¬£{df["price"].median():,.0f}')
axes[0, 0].set_xlabel('Precio (¬£)', fontsize=12)
axes[0, 0].set_ylabel('Frecuencia', fontsize=12)
axes[0, 0].set_title('Histograma del Precio', fontsize=14, fontweight='bold')
axes[0, 0].legend(fontsize=10)
axes[0, 0].grid(True, alpha=0.3)

# 2. Boxplot completo
bp = axes[0, 1].boxplot(df['price'], vert=True, patch_artist=True, 
                        boxprops=dict(facecolor='lightblue', color='black'),
                        medianprops=dict(color='red', linewidth=2),
                        whiskerprops=dict(color='black'),
                        capprops=dict(color='black'))
axes[0, 1].set_ylabel('Precio (¬£)', fontsize=12)
axes[0, 1].set_title('Boxplot del Precio (con outliers)', fontsize=14, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3, axis='y')

# 3. Boxplot sin outliers extremos (para mejor visualizaci√≥n)
filtered_price = df[(df['price'] >= q1 - 1.5*iqr) & (df['price'] <= q3 + 1.5*iqr)]['price']
bp2 = axes[0, 2].boxplot(filtered_price, vert=True, patch_artist=True,
                         boxprops=dict(facecolor='lightgreen', color='black'),
                         medianprops=dict(color='red', linewidth=2))
axes[0, 2].set_ylabel('Precio (¬£)', fontsize=12)
axes[0, 2].set_title('Boxplot sin Outliers Extremos', fontsize=14, fontweight='bold')
axes[0, 2].grid(True, alpha=0.3, axis='y')

# 4. Distribuci√≥n acumulada
axes[1, 0].hist(df['price'], bins=100, cumulative=True, density=True, 
                alpha=0.7, color='purple', edgecolor='black')
axes[1, 0].set_xlabel('Precio (¬£)', fontsize=12)
axes[1, 0].set_ylabel('Probabilidad Acumulada', fontsize=12)
axes[1, 0].set_title('Funci√≥n de Distribuci√≥n Acumulada', fontsize=14, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3)

# 5. Q-Q Plot (normalidad)
from scipy.stats import probplot
probplot(df['price'], dist="norm", plot=axes[1, 1])
axes[1, 1].set_title('Q-Q Plot (Test de Normalidad)', fontsize=14, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3)

# 6. Densidad (KDE)
df['price'].plot(kind='density', ax=axes[1, 2], color='darkblue', linewidth=2)
axes[1, 2].axvline(df['price'].mean(), color='red', linestyle='--', linewidth=2, label='Media')
axes[1, 2].axvline(df['price'].median(), color='green', linestyle='--', linewidth=2, label='Mediana')
axes[1, 2].set_xlabel('Precio (¬£)', fontsize=12)
axes[1, 2].set_ylabel('Densidad', fontsize=12)
axes[1, 2].set_title('Densidad de Probabilidad (KDE)', fontsize=14, fontweight='bold')
axes[1, 2].legend()
axes[1, 2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

### 3.6 Matriz de Correlaci√≥n

````python
print("=" * 80)
print("üîó AN√ÅLISIS DE CORRELACIONES ENTRE VARIABLES NUM√âRICAS")
print("=" * 80)

# Seleccionar solo variables num√©ricas
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_df = df[numeric_cols]

print(f"\nVariables num√©ricas analizadas: {numeric_cols}")
print(f"Total: {len(numeric_cols)} variables\n")

# Calcular matriz de correlaci√≥n
correlation_matrix = numeric_df.corr()

# Visualizar matriz de correlaci√≥n
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))  # M√°scara para mitad superior
sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1)
plt.title('Matriz de Correlaci√≥n - Variables Num√©ricas', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

# An√°lisis espec√≠fico de correlaci√≥n con el precio
print("\nüìà CORRELACI√ìN CON EL PRECIO (TARGET):")
print("=" * 60)
price_correlation = correlation_matrix['price'].sort_values(ascending=False)

for var, corr in price_correlation.items():
    if var != 'price':
        # Interpretar fuerza de correlaci√≥n
        if abs(corr) >= 0.7:
            strength = "MUY FUERTE"
            icon = "üî¥"
        elif abs(corr) >= 0.5:
            strength = "FUERTE"
            icon = "üü†"
        elif abs(corr) >= 0.3:
            strength = "MODERADA"
            icon = "üü°"
        elif abs(corr) >= 0.1:
            strength = "D√âBIL"
            icon = "üü¢"
        else:
            strength = "MUY D√âBIL"
            icon = "‚ö™"
        
        direction = "positiva" if corr > 0 else "negativa"
        print(f"{icon} {var:15s}: {corr:+.3f}  ‚Üí  Correlaci√≥n {strength} {direction}")

print("=" * 60)

# Interpretaciones clave
print("\nüí° INTERPRETACIONES CLAVE:")
print("\n1. Correlaciones POSITIVAS con el precio:")
for var, corr in price_correlation.items():
    if var != 'price' and corr > 0.3:
        print(f"   ‚Ä¢ {var}: {corr:.3f} - A mayor {var}, mayor precio")

print("\n2. Correlaciones NEGATIVAS con el precio:")
for var, corr in price_correlation.items():
    if var != 'price' and corr < -0.1:
        print(f"   ‚Ä¢ {var}: {corr:.3f} - A mayor {var}, menor precio")

# Multicolinealidad entre features
print("\n\n‚ö†Ô∏è AN√ÅLISIS DE MULTICOLINEALIDAD (correlaciones fuertes entre features):")
print("=" * 60)
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) >= 0.5:
            var1 = correlation_matrix.columns[i]
            var2 = correlation_matrix.columns[j]
            corr_value = correlation_matrix.iloc[i, j]
            if var1 != 'price' and var2 != 'price':
                high_corr_pairs.append((var1, var2, corr_value))

if high_corr_pairs:
    for var1, var2, corr in high_corr_pairs:
        print(f"   ‚Ä¢ {var1} ‚Üî {var2}: {corr:.3f}")
    print("\n   ‚ö†Ô∏è Estas correlaciones pueden causar multicolinealidad en regresi√≥n lineal")
else:
    print("   ‚úÖ No se detectaron correlaciones fuertes entre features (excluyendo price)")
````

### 4.1 An√°lisis y Manejo de Outliers

````python
print("=" * 80)
print("üîç AN√ÅLISIS DETALLADO DE OUTLIERS EN EL PRECIO")
print("=" * 80)

# M√©todo IQR para detectar outliers
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar outliers
outliers_mask = (df['price'] < lower_bound) | (df['price'] > upper_bound)
outliers_df = df[outliers_mask]
normal_df = df[~outliers_mask]

print(f"\nüìä Estad√≠sticas de Outliers:")
print(f"   Total de outliers: {len(outliers_df)} ({len(outliers_df)/len(df)*100:.2f}%)")
print(f"   Outliers inferiores: {len(df[df['price'] < lower_bound])}")
print(f"   Outliers superiores: {len(df[df['price'] > upper_bound])}")
print(f"\n   L√≠mite inferior: ¬£{lower_bound:,.2f}")
print(f"   L√≠mite superior: ¬£{upper_bound:,.2f}")
print(f"\n   Precio m√≠nimo outlier: ¬£{outliers_df['price'].min():,.2f}")
print(f"   Precio m√°ximo outlier: ¬£{outliers_df['price'].max():,.2f}")

# An√°lisis de outliers superiores (m√°s interesantes)
print(f"\nüîç An√°lisis de los 10 veh√≠culos M√ÅS CAROS:")
top_expensive = df.nlargest(10, 'price')[['model', 'year', 'price', 'mileage', 'transmission', 'fuelType', 'engineSize']]
display(top_expensive.style.background_gradient(cmap='Reds', subset=['price']))

# Visualizaci√≥n de outliers
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('An√°lisis Visual de Outliers', fontsize=16, fontweight='bold')

# 1. Boxplot con outliers marcados
axes[0].boxplot(df['price'], vert=True, patch_artist=True,
                boxprops=dict(facecolor='lightblue'),
                medianprops=dict(color='red', linewidth=2))
axes[0].axhline(upper_bound, color='orange', linestyle='--', linewidth=2, label=f'Upper bound: ¬£{upper_bound:,.0f}')
axes[0].axhline(lower_bound, color='orange', linestyle='--', linewidth=2, label=f'Lower bound: ¬£{lower_bound:,.0f}')
axes[0].set_ylabel('Precio (¬£)', fontsize=12, fontweight='bold')
axes[0].set_title('Boxplot con L√≠mites de Outliers', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3, axis='y')

# 2. Histograma comparativo
axes[1].hist([normal_df['price'], outliers_df['price']], bins=50, 
             label=['Normal', 'Outliers'], color=['skyblue', 'salmon'], alpha=0.7, edgecolor='black')
axes[1].set_xlabel('Precio (¬£)', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Frecuencia', fontsize=12, fontweight='bold')
axes[1].set_title('Distribuci√≥n: Normal vs Outliers', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# 3. Scatter: A√±o vs Precio (outliers destacados)
axes[2].scatter(normal_df['year'], normal_df['price'], alpha=0.4, s=20, color='blue', label='Normal')
axes[2].scatter(outliers_df['year'], outliers_df['price'], alpha=0.8, s=50, color='red', 
                marker='X', edgecolor='black', linewidth=1, label='Outliers')
axes[2].set_xlabel('A√±o', fontsize=12, fontweight='bold')
axes[2].set_ylabel('Precio (¬£)', fontsize=12, fontweight='bold')
axes[2].set_title('Outliers en Relaci√≥n A√±o-Precio', fontsize=14, fontweight='bold')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "=" * 80)

# Decisi√≥n sobre outliers
print("\nüí° DECISI√ìN SOBRE OUTLIERS:")
print("\n   Para este proyecto, MANTENDREMOS los outliers por las siguientes razones:")
print("   1. Son veh√≠culos leg√≠timos (Serie 7, X7, M Series de alta gama)")
print("   2. Representan un segmento importante del mercado BMW")
print("   3. Algoritmos como Random Forest y XGBoost son robustos a outliers")
print("   4. Eliminarlos sesgaria el modelo hacia veh√≠culos de gama media/baja")
print("\n   Alternativa: Usar algoritmos robustos y validaci√≥n cuidadosa")
print("=" * 80)
````

### 4.2 Feature Engineering - Creaci√≥n de Nuevas Variables

````python
print("=" * 80)
print("üîß FEATURE ENGINEERING - CREACI√ìN DE NUEVAS VARIABLES")
print("=" * 80)

# Crear una copia del dataframe para no modificar el original
df_processed = df.copy()

# 1. AGE: Edad del veh√≠culo (m√°s intuitivo que 'year')
current_year = 2020  # Asumimos que los datos son hasta 2020
df_processed['age'] = current_year - df_processed['year']

print("\n1Ô∏è‚É£ Variable 'age' (edad del veh√≠culo):")
print(f"   Rango: {df_processed['age'].min()} - {df_processed['age'].max()} a√±os")
print(f"   Media: {df_processed['age'].mean():.2f} a√±os")

# 2. MILEAGE_PER_YEAR: Kilometraje promedio anual
df_processed['mileage_per_year'] = df_processed['mileage'] / (df_processed['age'] + 1)  # +1 para evitar divisi√≥n por 0

print("\n2Ô∏è‚É£ Variable 'mileage_per_year' (km anuales):")
print(f"   Rango: {df_processed['mileage_per_year'].min():.0f} - {df_processed['mileage_per_year'].max():.0f} km/a√±o")
print(f"   Media: {df_processed['mileage_per_year'].mean():.0f} km/a√±o")

# 3. IS_LUXURY: Variable binaria para modelos de lujo
luxury_models = ['7 Series', 'X7', 'M5', 'M4', 'M3', 'M2', '8 Series', 'i8']
df_processed['is_luxury'] = df_processed['model'].apply(lambda x: 1 if x.strip() in luxury_models else 0)

print("\n3Ô∏è‚É£ Variable 'is_luxury' (modelo de lujo):")
print(f"   Veh√≠culos de lujo: {df_processed['is_luxury'].sum()} ({df_processed['is_luxury'].mean()*100:.2f}%)")

# 4. IS_AUTOMATIC: Variable binaria para transmisi√≥n autom√°tica
df_processed['is_automatic'] = df_processed['transmission'].apply(lambda x: 1 if x == 'Automatic' else 0)

print("\n4Ô∏è‚É£ Variable 'is_automatic' (transmisi√≥n autom√°tica):")
print(f"   Veh√≠culos autom√°ticos: {df_processed['is_automatic'].sum()} ({df_processed['is_automatic'].mean()*100:.2f}%)")

# 5. IS_DIESEL: Variable binaria para combustible di√©sel
df_processed['is_diesel'] = df_processed['fuelType'].apply(lambda x: 1 if x == 'Diesel' else 0)

print("\n5Ô∏è‚É£ Variable 'is_diesel' (combustible di√©sel):")
print(f"   Veh√≠culos di√©sel: {df_processed['is_diesel'].sum()} ({df_processed['is_diesel'].mean()*100:.2f}%)")

# 6. ENGINE_CATEGORY: Categor√≠a de tama√±o de motor
def categorize_engine(size):
    if size == 0:
        return 'electric'
    elif size < 1.5:
        return 'small'
    elif size < 2.5:
        return 'medium'
    elif size < 4.0:
        return 'large'
    else:
        return 'very_large'

df_processed['engine_category'] = df_processed['engineSize'].apply(categorize_engine)

print("\n6Ô∏è‚É£ Variable 'engine_category' (categor√≠a de motor):")
print(df_processed['engine_category'].value_counts())

# 7. PRICE_PER_YEAR: Precio por a√±o de vida restante (depreciaci√≥n relativa)
expected_life = 15  # Vida √∫til esperada de un veh√≠culo
df_processed['years_remaining'] = np.maximum(expected_life - df_processed['age'], 1)
df_processed['price_per_year'] = df_processed['price'] / df_processed['years_remaining']

print("\n7Ô∏è‚É£ Variable 'price_per_year' (valor por a√±o restante):")
print(f"   Media: ¬£{df_processed['price_per_year'].mean():.2f}/a√±o")

# Resumen de nuevas variables
print("\n" + "=" * 80)
print("üìä RESUMEN DE FEATURES CREADAS:")
print("=" * 80)
print(f"\nVariables originales: {df.shape[1]}")
print(f"Nuevas variables creadas: {df_processed.shape[1] - df.shape[1]}")
print(f"Total de variables: {df_processed.shape[1]}")
print("\nNuevas variables num√©ricas: age, mileage_per_year, price_per_year")
print("Nuevas variables binarias: is_luxury, is_automatic, is_diesel")
print("Nuevas variables categ√≥ricas: engine_category")
print("=" * 80)

# Visualizar correlaci√≥n de nuevas features con el precio
new_features = ['age', 'mileage_per_year', 'is_luxury', 'is_automatic', 'is_diesel', 'price_per_year']
new_corr = df_processed[new_features + ['price']].corr()['price'].drop('price').sort_values(ascending=False)

print("\nüîó CORRELACI√ìN DE NUEVAS FEATURES CON EL PRECIO:")
print("=" * 60)
for feat, corr in new_corr.items():
    print(f"   {feat:20s}: {corr:+.3f}")
print("=" * 60)
````

### 4.3 Selecci√≥n de Features para el Modelo

````python
print("=" * 80)
print("üéØ SELECCI√ìN DE FEATURES PARA EL MODELO")
print("=" * 80)

# Features que usaremos (excluyendo algunas redundantes)
features_to_use = [
    # Variables originales (num√©ricas)
    'year',
    'mileage',
    'tax',
    'mpg',
    'engineSize',
    
    # Variables originales (categ√≥ricas)
    'model',
    'transmission',
    'fuelType',
    
    # Nuevas features creadas
    'age',
    'mileage_per_year',
    'is_luxury',
    'is_automatic',
    'is_diesel'
]

# Nota: Eliminamos 'price_per_year' y 'years_remaining' porque usan el target en su c√°lculo (data leakage)

# Variable objetivo
target = 'price'

# Crear dataset final
X = df_processed[features_to_use].copy()
y = df_processed[target].copy()

print(f"\n‚úÖ Features seleccionadas: {len(features_to_use)}")
print(f"\nüìã Lista de features:")
for i, feat in enumerate(features_to_use, 1):
    feat_type = X[feat].dtype
    print(f"   {i:2d}. {feat:20s} (tipo: {feat_type})")

print(f"\nüéØ Variable objetivo: {target}")
print(f"\nüìä Dimensiones finales:")
print(f"   X shape: {X.shape}")
print(f"   y shape: {y.shape}")
print("=" * 80)
````

### 4.4 Encoding de Variables Categ√≥ricas

### 5.2 An√°lisis Detallado de Cada Algoritmo (P3, P4, M1, M2)

#### 1. Regresi√≥n Lineal - Baseline Simple

**¬øC√≥mo funciona?**

La regresi√≥n lineal modela la relaci√≥n entre las variables predictoras (X) y el target (y) mediante una ecuaci√≥n lineal:

$$\hat{y} = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_nx_n$$

El algoritmo encuentra los coeficientes (Œ≤) que minimizan el error cuadr√°tico medio (MSE):

$$MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$

**Proceso de aprendizaje:**
1. Inicializa coeficientes aleatoriamente
2. Calcula predicciones con la ecuaci√≥n lineal
3. Calcula el error (MSE)
4. Ajusta coeficientes para minimizar error (usando m√≠nimos cuadrados ordinarios)
5. Repite hasta convergencia

**Ventajas:**
- Extremadamente r√°pido de entrenar y predecir
- Altamente interpretable: cada coeficiente muestra el impacto directo de una variable
- No requiere tuning de hiperpar√°metros
- Funciona bien cuando las relaciones son aproximadamente lineales
- Excelente como modelo baseline

**Desventajas:**
- Asume relaci√≥n lineal (limitaci√≥n fuerte)
- Sensible a outliers (pueden sesgar la l√≠nea de regresi√≥n)
- Sufre de multicolinealidad (features correlacionados)
- No captura interacciones complejas entre variables
- Puede hacer underfitting en problemas complejos

**Hiperpar√°metros:**
- Ninguno (modelo no param√©trico)

**Por qu√© lo usamos:**
- Establecer un **baseline de desempe√±o m√≠nimo**
- Comparar modelos m√°s complejos contra este simple
- Validar que el problema no es trivialmente lineal

**Aplicabilidad a nuestro problema:**
- Moderada: La depreciaci√≥n de veh√≠culos NO es perfectamente lineal
- Esperamos R¬≤ ‚âà 0.70-0.75 (decente pero mejorable)

#### 2. Ridge Regression - Baseline con Regularizaci√≥n

**¬øC√≥mo funciona?**

Ridge es una variante de regresi√≥n lineal que a√±ade un t√©rmino de **regularizaci√≥n L2** para prevenir overfitting:

$$\text{Minimize: } MSE + \alpha\sum_{j=1}^{p}\beta_j^2$$

Donde Œ± (alpha) controla la fuerza de la regularizaci√≥n.

**Proceso de aprendizaje:**
1. Similar a regresi√≥n lineal, pero con penalizaci√≥n
2. Los coeficientes grandes son penalizados
3. Reduce la magnitud de coeficientes pero no los elimina (los encoge hacia 0)
4. Encuentra un balance entre ajuste y complejidad

**Ventajas:**
- Previene overfitting mejor que regresi√≥n lineal simple
- Maneja multicolinealidad mejor (features correlacionados)
- M√°s estable num√©ricamente
- Sigue siendo interpretable
- Buen desempe√±o con muchas features

**Desventajas:**
- Requiere tuning del hiperpar√°metro Œ±
- No realiza feature selection (todos los coeficientes ‚â† 0)
- Sigue asumiendo relaciones lineales
- Puede hacer underfitting si Œ± es muy alto

**Hiperpar√°metros clave:**
- **alpha (Œ±):** Fuerza de regularizaci√≥n
  - Œ± = 0 ‚Üí Regresi√≥n lineal normal
  - Œ± peque√±o (0.01-1) ‚Üí Poca regularizaci√≥n
  - Œ± grande (10-1000) ‚Üí Alta regularizaci√≥n, coeficientes muy peque√±os

**Por qu√© lo usamos:**
- Mejora sobre regresi√≥n lineal manejando multicolinealidad
- Baseline m√°s robusto que regresi√≥n simple
- Comparar efecto de la regularizaci√≥n

**Aplicabilidad a nuestro problema:**
- Buena: Nuestro dataset tiene algunas correlaciones entre features
- Esperamos R¬≤ ‚âà 0.72-0.77 (ligera mejora sobre lineal)

#### 3. Random Forest - Ensemble Robusto

**¬øC√≥mo funciona?**

Random Forest es un **ensemble** de m√∫ltiples √°rboles de decisi√≥n entrenados de forma independiente:

1. **Bagging:** Cada √°rbol se entrena con una muestra aleatoria del dataset (con reemplazo)
2. **Feature randomness:** En cada split, solo se consideran un subconjunto aleatorio de features
3. **Predicci√≥n:** Promedio de las predicciones de todos los √°rboles

**Proceso de aprendizaje de cada √°rbol:**
1. Seleccionar muestra bootstrap (aleatoria con reemplazo)
2. En cada nodo, elegir aleatoriamente ‚àöp features (p = total de features)
3. Encontrar el mejor split entre esas features (minimizando MSE)
4. Repetir recursivamente hasta max_depth o min_samples_leaf
5. Predicci√≥n final = promedio de N √°rboles

**Ventajas:**
- **Muy robusto:** Reduce overfitting mediante averaging
- Captura relaciones no lineales autom√°ticamente
- No requiere escalado de features
- Maneja outliers bien
- Proporciona **feature importance** (interpretabilidad parcial)
- Funciona bien "out of the box" sin mucho tuning
- Paralelizable (√°rboles independientes)

**Desventajas:**
- Menos interpretable que un solo √°rbol
- M√°s lento que modelos lineales (N √°rboles)
- Puede requerir mucha memoria con muchos √°rboles
- Tiende a sobreajustarse en datasets muy ruidosos

**Hiperpar√°metros clave:**
- **n_estimators:** N√∫mero de √°rboles (t√≠picamente 100-500)
  - M√°s √°rboles ‚Üí mejor desempe√±o, pero m√°s lento
- **max_depth:** Profundidad m√°xima de cada √°rbol (t√≠picamente 10-30)
  - Muy profundo ‚Üí overfitting
  - Muy poco ‚Üí underfitting
- **min_samples_split:** M√≠nimo de muestras para dividir un nodo (t√≠picamente 2-10)
- **min_samples_leaf:** M√≠nimo de muestras en una hoja (t√≠picamente 1-5)
- **max_features:** Features a considerar en cada split (t√≠picamente 'sqrt' o 'log2')

**Por qu√© lo usamos:**
- **Uno de los mejores algoritmos** para datos tabulares
- Ganador frecuente en competiciones (junto con Gradient Boosting)
- Balance perfecto entre desempe√±o y facilidad de uso

**Aplicabilidad a nuestro problema:**
- Excelente: Perfecto para capturar no linealidad en depreciaci√≥n
- Esperamos R¬≤ ‚âà 0.82-0.87 (muy bueno)

#### 4. XGBoost - Gradient Boosting Optimizado

**¬øC√≥mo funciona?**

XGBoost (eXtreme Gradient Boosting) construye √°rboles **secuencialmente**, donde cada √°rbol intenta corregir los errores del anterior:

1. **√Årbol 1:** Predice el target
2. **√Årbol 2:** Predice los **residuos** (errores) del √°rbol 1
3. **√Årbol 3:** Predice los residuos del √°rbol 2
4. ...
5. **Predicci√≥n final:** Suma ponderada de todos los √°rboles

$$\hat{y} = \sum_{k=1}^{K} f_k(x)$$

Donde cada $f_k$ es un √°rbol que minimiza:

$$L = \sum_{i=1}^{n}l(y_i, \hat{y}_i) + \sum_{k=1}^{K}\Omega(f_k)$$

(Loss function + Regularizaci√≥n)

**Proceso de aprendizaje:**
1. Inicializar predicci√≥n (t√≠picamente media del target)
2. Para cada √°rbol k:
   - Calcular gradientes (‚àÇLoss/‚àÇpredicci√≥n)
   - Entrenar √°rbol para predecir estos gradientes
   - A√±adir √°rbol con learning_rate (Œ∑)
3. Repetir hasta n_estimators o convergencia

**Ventajas:**
- **Extremadamente preciso:** Uno de los mejores algoritmos para datos tabulares
- Regularizaci√≥n incorporada (previene overfitting)
- Maneja relaciones complejas y no lineales
- Eficiente computacionalmente (implementaci√≥n optimizada)
- Feature importance incorporado
- Maneja missing values nativamente
- **Ganador frecuente en Kaggle**

**Desventajas:**
- Requiere **tuning cuidadoso** de hiperpar√°metros
- Puede hacer overfitting si no se configura bien
- M√°s lento de entrenar que Random Forest (secuencial)
- Menos interpretable (muchos √°rboles secuenciales)
- Sensible a outliers en algunos casos

**Hiperpar√°metros clave:**
- **n_estimators:** N√∫mero de √°rboles (t√≠picamente 100-1000)
- **learning_rate (eta):** Tasa de aprendizaje (t√≠picamente 0.01-0.3)
  - Bajo (0.01) ‚Üí m√°s √°rboles necesarios, m√°s lento, mejor generalizaci√≥n
  - Alto (0.3) ‚Üí menos √°rboles, m√°s r√°pido, riesgo de overfitting
- **max_depth:** Profundidad de √°rboles (t√≠picamente 3-10)
  - M√°s bajo que Random Forest (√°rboles m√°s simples)
- **subsample:** Fracci√≥n de muestras por √°rbol (t√≠picamente 0.8-1.0)
- **colsample_bytree:** Fracci√≥n de features por √°rbol (t√≠picamente 0.8-1.0)
- **reg_alpha (L1), reg_lambda (L2):** Regularizaci√≥n

**Por qu√© lo usamos:**
- **Estado del arte** en problemas de regresi√≥n tabulares
- Esperamos que sea nuestro mejor modelo
- Industria lo usa ampliamente (Airbnb, Uber, etc.)

**Aplicabilidad a nuestro problema:**
- Excelente: Perfecto para capturar patrones complejos en precios
- Esperamos R¬≤ ‚âà 0.85-0.90 (el m√°s alto)

#### 5. Support Vector Regression (SVR) con kernel RBF

**¬øC√≥mo funciona?**

SVR encuentra una funci√≥n que se ajusta a los datos dentro de un margen de tolerancia Œµ (epsilon):

- **Objetivo:** Encontrar $f(x) = w^Tx + b$ tal que la mayor√≠a de los puntos est√©n dentro de un "tubo" de ancho 2Œµ
- Puntos fuera del tubo son penalizados
- **Kernel Trick:** Mapea datos a un espacio de mayor dimensi√≥n donde pueden ser linealmente separables

**Kernel RBF (Radial Basis Function):**

$$K(x_i, x_j) = \exp(-\gamma||x_i - x_j||^2)$$

Permite capturar relaciones no lineales mapeando a un espacio infinito-dimensional.

**Proceso de aprendizaje:**
1. Mapear datos al espacio de features (mediante kernel)
2. Encontrar hiperplano √≥ptimo en ese espacio
3. Minimizar: $\frac{1}{2}||w||^2 + C\sum\xi_i$ (regularizaci√≥n + errores)
4. Usar solo "support vectors" (puntos cr√≠ticos) para la predicci√≥n

**Ventajas:**
- Efectivo en espacios de alta dimensi√≥n
- Captura relaciones no lineales con kernels
- Robusto cuando hay clara separaci√≥n entre regiones
- Usa solo support vectors (eficiente en memoria despu√©s del entrenamiento)

**Desventajas:**
- **Costoso computacionalmente** con datasets grandes (O(n¬≤) a O(n¬≥))
- Muy sensible a hiperpar√°metros (C, gamma, epsilon)
- **Requiere escalado** de features (obligatorio)
- Dif√≠cil de interpretar con kernels no lineales
- No proporciona feature importance directamente

**Hiperpar√°metros clave:**
- **C:** Par√°metro de regularizaci√≥n (t√≠picamente 0.1-100)
  - Bajo ‚Üí margen amplio, m√°s generalizaci√≥n, m√°s errores permitidos
  - Alto ‚Üí margen estrecho, menos errores, riesgo de overfitting
- **gamma:** Para kernel RBF (t√≠picamente 0.001-1.0)
  - Bajo ‚Üí influencia amplia (smooth)
  - Alto ‚Üí influencia local (riesgo de overfitting)
- **epsilon (Œµ):** Ancho del tubo de tolerancia (t√≠picamente 0.01-1.0)
  - Define cu√°nto error es aceptable sin penalizaci√≥n

**Por qu√© lo usamos:**
- Representar paradigma de **optimizaci√≥n + kernel**
- Comparar con m√©todos basados en √°rboles
- Validar si mapeo no lineal ayuda

**Aplicabilidad a nuestro problema:**
- Buena: Puede capturar no linealidad, pero costoso con 10k+ ejemplos
- Esperamos R¬≤ ‚âà 0.78-0.83 (bueno pero no el mejor)

#### 6. Neural Network (Multi-Layer Perceptron)

**¬øC√≥mo funciona?**

Una red neuronal MLP consiste en capas de neuronas conectadas:

**Arquitectura:**
- **Capa de entrada:** Recibe los 18 features
- **Capas ocultas:** 1-3 capas con 50-200 neuronas cada una
- **Capa de salida:** 1 neurona (predicci√≥n del precio)

**Matem√°tica de una neurona:**

$$h = \sigma(w^Tx + b)$$

Donde:
- $w$ = pesos
- $x$ = entrada
- $b$ = sesgo (bias)
- $\sigma$ = funci√≥n de activaci√≥n (ReLU, Sigmoid, Tanh)

**Proceso de aprendizaje (Backpropagation):**
1. **Forward pass:** Propagar entrada hacia adelante, calcular predicci√≥n
2. **Calcular loss:** $MSE = \frac{1}{n}\sum(y - \hat{y})^2$
3. **Backward pass:** Calcular gradientes (‚àÇLoss/‚àÇweights) mediante regla de la cadena
4. **Actualizar pesos:** $w = w - \eta \cdot \nabla Loss$ (Gradient Descent)
5. Repetir por m√∫ltiples epochs

**Ventajas:**
- Captura relaciones **extremadamente complejas y no lineales**
- Aprende autom√°ticamente representaciones de features
- Flexible: ajustable a casi cualquier problema
- Escalable a grandes datasets (con GPU)
- Estado del arte en visi√≥n, NLP, etc.

**Desventajas:**
- Requiere **muchos datos** (t√≠picamente miles/millones)
- Muchos hiperpar√°metros para tunear
- **"Black box":** Muy dif√≠cil de interpretar
- Costoso computacionalmente (GPU recomendada)
- Propenso a overfitting sin regularizaci√≥n
- Puede quedarse en m√≠nimos locales

**Hiperpar√°metros clave:**
- **hidden_layer_sizes:** Arquitectura (ej: (100, 50) = 2 capas)
- **activation:** Funci√≥n de activaci√≥n ('relu', 'tanh', 'sigmoid')
  - ReLU (Rectified Linear Unit) es la m√°s com√∫n
- **alpha:** Regularizaci√≥n L2 (t√≠picamente 0.0001-0.01)
- **learning_rate_init:** Tasa de aprendizaje inicial (t√≠picamente 0.001-0.01)
- **max_iter:** N√∫mero de epochs (t√≠picamente 200-1000)
- **batch_size:** Tama√±o de mini-batch (t√≠picamente 32-256)

**Por qu√© lo usamos:**
- Representar **paradigma de deep learning**
- Comparar con m√©todos tradicionales
- Ver si complejidad adicional ayuda

**Aplicabilidad a nuestro problema:**
- Moderada: Dataset es relativamente peque√±o para deep learning
- Esperamos R¬≤ ‚âà 0.80-0.85 (bueno, pero puede ser superado por XGBoost con menos esfuerzo)

### 6.1 Preparaci√≥n para el Entrenamiento


````python
# Funci√≥n para evaluar modelos
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    """
    Eval√∫a un modelo de regresi√≥n y retorna m√©tricas detalladas
    """
    import time
    
    # Medir tiempo de entrenamiento
    start_time = time.time()
    model.fit(X_train, y_train)
    training_time = time.time() - start_time
    
    # Predicciones
    start_time = time.time()
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    prediction_time = time.time() - start_time
    
    # M√©tricas en train
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    train_r2 = r2_score(y_train, y_train_pred)
    train_mape = np.mean(np.abs((y_train - y_train_pred) / y_train)) * 100
    
    # M√©tricas en test
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    test_r2 = r2_score(y_test, y_test_pred)
    test_mape = np.mean(np.abs((y_test - y_test_pred) / y_test)) * 100
    
    # Resultados
    results = {
        'model_name': model_name,
        'model_object': model,
        'train_mae': train_mae,
        'train_rmse': train_rmse,
        'train_r2': train_r2,
        'train_mape': train_mape,
        'test_mae': test_mae,
        'test_rmse': test_rmse,
        'test_r2': test_r2,
        'test_mape': test_mape,
        'training_time': training_time,
        'prediction_time': prediction_time,
        'y_train_pred': y_train_pred,
        'y_test_pred': y_test_pred
    }
    
    # Imprimir resultados
    print(f"\n{'='*80}")
    print(f"üìä RESULTADOS: {model_name}")
    print(f"{'='*80}")
    print(f"\nüîπ Training Set:")
    print(f"   MAE:  ¬£{train_mae:,.2f}")
    print(f"   RMSE: ¬£{train_rmse:,.2f}")
    print(f"   R¬≤:   {train_r2:.4f}")
    print(f"   MAPE: {train_mape:.2f}%")
    
    print(f"\nüîπ Test Set:")
    print(f"   MAE:  ¬£{test_mae:,.2f}")
    print(f"   RMSE: ¬£{test_rmse:,.2f}")
    print(f"   R¬≤:   {test_r2:.4f}")
    print(f"   MAPE: {test_mape:.2f}%")
    
    print(f"\n‚è±Ô∏è  Tiempos:")
    print(f"   Entrenamiento: {training_time:.3f}s")
    print(f"   Predicci√≥n:    {prediction_time:.3f}s")
    
    # An√°lisis de overfitting/underfitting
    r2_diff = train_r2 - test_r2
    if r2_diff > 0.1:
        print(f"\n‚ö†Ô∏è  OVERFITTING detectado (Œî R¬≤ = {r2_diff:.4f})")
        print("    El modelo se ajusta demasiado al training set")
    elif test_r2 < 0.75:
        print(f"\n‚ö†Ô∏è  UNDERFITTING detectado (R¬≤ test = {test_r2:.4f})")
        print("    El modelo no captura bien los patrones")
    else:
        print(f"\n‚úÖ Buen balance (Œî R¬≤ = {r2_diff:.4f})")
    
    print(f"{'='*80}\n")
    
    return results

# Diccionario para almacenar todos los resultados
all_results = {}

print("‚úÖ Funci√≥n de evaluaci√≥n definida")
print("üì¶ Diccionario de resultados inicializado")
````

### 7.1 Hyperparameter Tuning - XGBoost (Optimizaci√≥n Autom√°tica)

````python
print("=" * 80)
print("üéØ HYPERPARAMETER TUNING - XGBOOST (Optimizaci√≥n mediante Grid Search)")
print("=" * 80)

if not XGBOOST_AVAILABLE:
    print("‚ö†Ô∏è XGBoost no disponible. Saltando tuning.")
else:
    from sklearn.model_selection import GridSearchCV
    
    # Grid de hiperpar√°metros a probar
    param_grid = {
        'n_estimators': [200, 300],
        'learning_rate': [0.05, 0.1],
        'max_depth': [5, 6, 7],
        'min_child_weight': [2, 3],
        'subsample': [0.8],
        'colsample_bytree': [0.8]
    }
    
    print("\nüîç Configuraci√≥n de Grid Search:")
    print(f"   ‚Ä¢ Par√°metros a optimizar: {list(param_grid.keys())}")
    print(f"   ‚Ä¢ Combinaciones totales: {np.prod([len(v) for v in param_grid.values()])}")
    print(f"   ‚Ä¢ CV folds: 3")
    print(f"   ‚Ä¢ M√©trica: R¬≤")
    
    # Crear modelo base
    xgb_base = xgb.XGBRegressor(
        random_state=42,
        n_jobs=-1,
        verbosity=0
    )
    
    # Grid Search con Cross-Validation
    print("\n‚è≥ Ejecutando Grid Search (esto puede tardar varios minutos)...")
    grid_search = GridSearchCV(
        estimator=xgb_base,
        param_grid=param_grid,
        cv=3,                    # 3-fold cross-validation
        scoring='r2',
        n_jobs=-1,
        verbose=1
    )
    
    grid_search.fit(X_train, y_train)
    
    print("\n‚úÖ Grid Search completado")
    print("\nüèÜ MEJORES HIPERPAR√ÅMETROS ENCONTRADOS:")
    print("=" * 60)
    for param, value in grid_search.best_params_.items():
        print(f"   ‚Ä¢ {param:20s}: {value}")
    
    print(f"\nüìä R¬≤ Score (CV): {grid_search.best_score_:.4f}")
    
    # Entrenar modelo final con mejores par√°metros
    xgb_tuned = grid_search.best_estimator_
    
    # Evaluar en test set
    xgb_tuned_results = evaluate_model(
        model=xgb_tuned,
        X_train=X_train,
        y_train=y_train,
        X_test=X_test,
        y_test=y_test,
        model_name="XGBoost (Tuned)"
    )
    
    all_results['XGBoost Tuned'] = xgb_tuned_results
    
    # Comparar con modelo original
    print("\nüìà COMPARACI√ìN: XGBoost Original vs Tuned")
    print("=" * 70)
    print(f"   {'M√©trica':<20s} {'Original':>15s} {'Tuned':>15s} {'Mejora':>15s}")
    print("   " + "-" * 70)
    
    original_r2 = all_results['XGBoost']['test_r2']
    tuned_r2 = xgb_tuned_results['test_r2']
    improvement_r2 = tuned_r2 - original_r2
    
    original_rmse = all_results['XGBoost']['test_rmse']
    tuned_rmse = xgb_tuned_results['test_rmse']
    improvement_rmse = original_rmse - tuned_rmse
    
    print(f"   {'R¬≤ Score':<20s} {original_r2:>15.4f} {tuned_r2:>15.4f} {improvement_r2:>+15.4f}")
    print(f"   {'RMSE':<20s} ¬£{original_rmse:>13,.2f} ¬£{tuned_rmse:>13,.2f} ¬£{improvement_rmse:>+13,.2f}")
    
    if improvement_r2 > 0:
        print(f"\n‚úÖ Tuning mejor√≥ el modelo en {improvement_r2:.4f} puntos de R¬≤")
    else:
        print(f"\n‚ö†Ô∏è El modelo original ya estaba bien calibrado")
````

## 9. Pruebas Autom√°ticas y Validaci√≥n

### Introducci√≥n

Esta secci√≥n cumple con: *Realizar pruebas autom√°ticas en una aplicaci√≥n de procesamiento de datos*.

Implementaremos pruebas autom√°ticas para validar:
1. Integridad de los datos
2. Rendimiento del modelo
3. Robustez ante diferentes escenarios
4. Calidad de las predicciones

### 9.1 Suite de Pruebas Autom√°ticas

````python
print("=" * 80)
print("üß™ SUITE DE PRUEBAS AUTOM√ÅTICAS (LO4)")
print("=" * 80)

# Clase para organizar las pruebas
class ModelTestSuite:
    """Suite de pruebas autom√°ticas para validar el modelo de ML"""
    
    def __init__(self, model, X_test, y_test, y_pred, model_name):
        self.model = model
        self.X_test = X_test
        self.y_test = y_test
        self.y_pred = y_pred
        self.model_name = model_name
        self.test_results = []
    
    def run_all_tests(self):
        """Ejecuta todas las pruebas"""
        print(f"\nüß™ Ejecutando pruebas para: {self.model_name}\n")
        
        self.test_data_integrity()
        self.test_prediction_range()
        self.test_model_performance()
        self.test_prediction_distribution()
        self.test_outlier_handling()
        self.test_feature_importance()
        self.test_prediction_consistency()
        
        # Resumen
        self.print_summary()
    
    def test_data_integrity(self):
        """Prueba 1: Integridad de datos"""
        test_name = "Test 1: Integridad de Datos"
        try:
            # Verificar que no hay NaN en predicciones
            assert not np.isnan(self.y_pred).any(), "Predicciones contienen NaN"
            
            # Verificar que no hay infinitos
            assert not np.isinf(self.y_pred).any(), "Predicciones contienen infinitos"
            
            # Verificar dimensiones
            assert len(self.y_pred) == len(self.y_test), "Dimensiones no coinciden"
            
            self.test_results.append((test_name, "‚úÖ PASS", "Datos v√°lidos"))
            print(f"‚úÖ {test_name}: PASS")
        except AssertionError as e:
            self.test_results.append((test_name, "‚ùå FAIL", str(e)))
            print(f"‚ùå {test_name}: FAIL - {e}")
    
    def test_prediction_range(self):
        """Prueba 2: Rango de predicciones"""
        test_name = "Test 2: Rango de Predicciones"
        try:
            # Las predicciones deben estar en un rango razonable
            min_pred = self.y_pred.min()
            max_pred = self.y_pred.max()
            
            # No deben ser negativas (precio no puede ser negativo)
            assert min_pred >= 0, f"Predicci√≥n negativa detectada: ¬£{min_pred:,.2f}"
            
            # No deben exceder un rango razonable (ej: ¬£200,000)
            assert max_pred < 200000, f"Predicci√≥n excesiva detectada: ¬£{max_pred:,.2f}"
            
            self.test_results.append((test_name, "‚úÖ PASS", 
                                     f"Rango: ¬£{min_pred:,.0f} - ¬£{max_pred:,.0f}"))
            print(f"‚úÖ {test_name}: PASS (Rango: ¬£{min_pred:,.0f} - ¬£{max_pred:,.0f})")
        except AssertionError as e:
            self.test_results.append((test_name, "‚ùå FAIL", str(e)))
            print(f"‚ùå {test_name}: FAIL - {e}")
    
    def test_model_performance(self):
        """Prueba 3: Desempe√±o m√≠nimo"""
        test_name = "Test 3: Desempe√±o M√≠nimo"
        try:
            r2 = r2_score(self.y_test, self.y_pred)
            mae = mean_absolute_error(self.y_test, self.y_pred)
            
            # R¬≤ debe ser al menos 0.70
            assert r2 >= 0.70, f"R¬≤ muy bajo: {r2:.4f} (m√≠nimo esperado: 0.70)"
            
            # MAE no debe exceder ¬£3,000
            assert mae <= 3000, f"MAE muy alto: ¬£{mae:,.2f} (m√°ximo esperado: ¬£3,000)"
            
            self.test_results.append((test_name, "‚úÖ PASS", 
                                     f"R¬≤={r2:.4f}, MAE=¬£{mae:,.0f}"))
            print(f"‚úÖ {test_name}: PASS (R¬≤={r2:.4f}, MAE=¬£{mae:,.0f})")
        except AssertionError as e:
            self.test_results.append((test_name, "‚ùå FAIL", str(e)))
            print(f"‚ùå {test_name}: FAIL - {e}")
    
    def test_prediction_distribution(self):
        """Prueba 4: Distribuci√≥n de predicciones"""
        test_name = "Test 4: Distribuci√≥n de Predicciones"
        try:
            # La distribuci√≥n de predicciones debe ser similar a la de valores reales
            pred_mean = self.y_pred.mean()
            real_mean = self.y_test.mean()
            
            # Diferencia de medias no debe ser >15%
            diff_pct = abs(pred_mean - real_mean) / real_mean * 100
            assert diff_pct < 15, f"Distribuciones muy diferentes: {diff_pct:.1f}% de diferencia"
            
            self.test_results.append((test_name, "‚úÖ PASS", 
                                     f"Diferencia de medias: {diff_pct:.1f}%"))
            print(f"‚úÖ {test_name}: PASS (Diferencia de medias: {diff_pct:.1f}%)")
        except AssertionError as e:
            self.test_results.append((test_name, "‚ùå FAIL", str(e)))
            print(f"‚ùå {test_name}: FAIL - {e}")
    
    def test_outlier_handling(self):
        """Prueba 5: Manejo de outliers"""
        test_name = "Test 5: Manejo de Outliers"
        try:
            # Identificar outliers en valores reales
            Q1 = self.y_test.quantile(0.25)
            Q3 = self.y_test.quantile(0.75)
            IQR = Q3 - Q1
            outliers_mask = (self.y_test < Q1 - 1.5*IQR) | (self.y_test > Q3 + 1.5*IQR)
            
            if outliers_mask.sum() > 0:
                # Calcular error en outliers
                outlier_mae = mean_absolute_error(
                    self.y_test[outliers_mask], 
                    self.y_pred[outliers_mask.values]
                )
                
                # Error en outliers no debe ser >2x el MAE general
                general_mae = mean_absolute_error(self.y_test, self.y_pred)
                assert outlier_mae < 2 * general_mae, \
                    f"Mal manejo de outliers: MAE={outlier_mae:.0f} (>2x general)"
                
                self.test_results.append((test_name, "‚úÖ PASS", 
                                         f"MAE outliers: ¬£{outlier_mae:,.0f}"))
                print(f"‚úÖ {test_name}: PASS (MAE outliers: ¬£{outlier_mae:,.0f})")
            else:
                self.test_results.append((test_name, "‚úÖ PASS", "No hay outliers"))
                print(f"‚úÖ {test_name}: PASS (No hay outliers en test set)")
        except AssertionError as e:
            self.test_results.append((test_name, "‚ùå FAIL", str(e)))
            print(f"‚ùå {test_name}: FAIL - {e}")
    
    def test_feature_importance(self):
        """Prueba 6: Feature importance (si aplica)"""
        test_name = "Test 6: Feature Importance"
        try:
            # Solo para modelos que tienen feature_importances_
            if hasattr(self.model, 'feature_importances_'):
                importances = self.model.feature_importances_
                
                # La suma debe ser ‚âà 1.0
                assert abs(importances.sum() - 1.0) < 0.01, \
                    f"Suma de importancias incorrecta: {importances.sum():.4f}"
                
                # Al menos 3 features deben tener importancia >0.05
                important_features = (importances > 0.05).sum()
                assert important_features >= 3, \
                    f"Muy pocas features importantes: {important_features}"
                
                self.test_results.append((test_name, "‚úÖ PASS", 
                                         f"{important_features} features importantes"))
                print(f"‚úÖ {test_name}: PASS ({important_features} features importantes)")
            else:
                self.test_results.append((test_name, "‚ö†Ô∏è SKIP", 
                                         "Modelo no tiene feature_importances_"))
                print(f"‚ö†Ô∏è {test_name}: SKIP (Modelo no aplicable)")
        except AssertionError as e:
            self.test_results.append((test_name, "‚ùå FAIL", str(e)))
            print(f"‚ùå {test_name}: FAIL - {e}")
    
    def test_prediction_consistency(self):
        """Prueba 7: Consistencia de predicciones"""
        test_name = "Test 7: Consistencia de Predicciones"
        try:
            # Predecir el mismo dato m√∫ltiples veces debe dar el mismo resultado
            sample_X = self.X_test.iloc[[0]]
            pred1 = self.model.predict(sample_X)[0]
            pred2 = self.model.predict(sample_X)[0]
            pred3 = self.model.predict(sample_X)[0]
            
            # Deben ser exactamente iguales
            assert pred1 == pred2 == pred3, "Predicciones inconsistentes para mismo input"
            
            self.test_results.append((test_name, "‚úÖ PASS", "Predicciones consistentes"))
            print(f"‚úÖ {test_name}: PASS (Predicciones consistentes)")
        except AssertionError as e:
            self.test_results.append((test_name, "‚ùå FAIL", str(e)))
            print(f"‚ùå {test_name}: FAIL - {e}")
    
    def print_summary(self):
        """Imprime resumen de pruebas"""
        print("\n" + "=" * 80)
        print(f"üìä RESUMEN DE PRUEBAS - {self.model_name}")
        print("=" * 80)
        
        passed = sum(1 for r in self.test_results if r[1] == "‚úÖ PASS")
        failed = sum(1 for r in self.test_results if r[1] == "‚ùå FAIL")
        skipped = sum(1 for r in self.test_results if r[1] == "‚ö†Ô∏è SKIP")
        total = len(self.test_results)
        
        print(f"\n   Total de pruebas: {total}")
        print(f"   ‚úÖ Pasadas: {passed}")
        print(f"   ‚ùå Fallidas: {failed}")
        print(f"   ‚ö†Ô∏è Saltadas: {skipped}")
        
        success_rate = (passed / (total - skipped)) * 100 if (total - skipped) > 0 else 0
        print(f"\n   Tasa de √©xito: {success_rate:.1f}%")
        
        if failed == 0:
            print("\n   üéâ TODAS LAS PRUEBAS PASARON")
        else:
            print(f"\n   ‚ö†Ô∏è HAY {failed} PRUEBA(S) FALLIDA(S)")
        
        print("=" * 80)

# Ejecutar pruebas para el mejor modelo
test_suite = ModelTestSuite(
    model=best_model_results['model_object'],
    X_test=X_test,
    y_test=y_test,
    y_pred=best_model_results['y_test_pred'],
    model_name=best_model_name
)

test_suite.run_all_tests()
````

# MODELO DE PREDICCI√ìN DE PRECIOS BMW - VERSI√ìN PRODUCCI√ìN


````python
import pandas as pd
import numpy as np
import pickle
from sklearn.preprocessing import StandardScaler
import xgboost as xgb

class BMWPricePredictor:
    """
    Predictor de precios de veh√≠culos BMW
    Modelo entrenado con XGBoost - R¬≤ Score: 0.88
    """
    
    def __init__(self, model=None, scaler_obj=None):
        """Cargar modelo y scaler pre-entrenados"""
        if model is None or scaler_obj is None:
            raise ValueError("Model and scaler objects must be provided")
        self.model = model
        self.scaler = scaler_obj
        
        # Mapeo de modelos a precio medio (target encoding)
        self.model_encoding = {
            '1 Series': 15000, '2 Series': 18000, '3 Series': 20000,
            '4 Series': 22000, '5 Series': 25000, '6 Series': 30000,
            '7 Series': 35000, '8 Series': 45000, 'M2': 40000,
            'M3': 45000, 'M4': 48000, 'M5': 52000, 'M8': 70000,
            'X1': 22000, 'X2': 24000, 'X3': 28000, 'X4': 32000,
            'X5': 38000, 'X6': 42000, 'X7': 50000, 'i3': 20000,
            'i4': 38000, 'i8': 60000
        }
    
    def preprocess(self, vehicle_data):
        """
        Preprocesar datos de entrada
        
        Args:
            vehicle_data (dict): Diccionario con caracter√≠sticas del veh√≠culo
                - model (str): Modelo BMW
                - year (int): A√±o de fabricaci√≥n
                - transmission (str): Manual/Automatic/Semi-Auto
                - mileage (int): Kilometraje
                - fuelType (str): Diesel/Petrol/Hybrid/Electric/Other
                - tax (float): Impuesto anual
                - mpg (float): Millas por gal√≥n
                - engineSize (float): Tama√±o del motor en litros
        
        Returns:
            np.array: Array con features procesadas listas para predicci√≥n
        """
        # Feature engineering
        current_year = 2024
        age = current_year - vehicle_data['year']
        mileage_per_year = vehicle_data['mileage'] / (age + 1)
        
        # Variables binarias
        luxury_models = ['7 Series', 'X7', 'M5', 'M4', 'M3', 'M2', '8 Series', 'i8']
        is_luxury = 1 if vehicle_data['model'] in luxury_models else 0
        is_automatic = 1 if vehicle_data['transmission'] == 'Automatic' else 0
        is_diesel = 1 if vehicle_data['fuelType'] == 'Diesel' else 0
        
        # Target encoding del modelo
        model_encoded = self.model_encoding.get(vehicle_data['model'], 25000)
        
        # One-hot encoding
        trans_manual = 1 if vehicle_data['transmission'] == 'Manual' else 0
        trans_semi = 1 if vehicle_data['transmission'] == 'Semi-Auto' else 0
        
        fuel_electric = 1 if vehicle_data['fuelType'] == 'Electric' else 0
        fuel_hybrid = 1 if vehicle_data['fuelType'] == 'Hybrid' else 0
        fuel_other = 1 if vehicle_data['fuelType'] == 'Other' else 0
        fuel_petrol = 1 if vehicle_data['fuelType'] == 'Petrol' else 0
        
        # Ensamblar features en el orden correcto
        features = [
            vehicle_data['year'],
            vehicle_data['mileage'],
            vehicle_data['tax'],
            vehicle_data['mpg'],
            vehicle_data['engineSize'],
            model_encoded,
            age,
            mileage_per_year,
            is_luxury,
            is_automatic,
            is_diesel,
            trans_manual,
            trans_semi,
            fuel_electric,
            fuel_hybrid,
            fuel_other,
            fuel_petrol
        ]
        
        return np.array(features).reshape(1, -1)
    
    def predict(self, vehicle_data):
        """
        Predecir precio de un veh√≠culo
        
        Args:
            vehicle_data (dict): Datos del veh√≠culo
        
        Returns:
            float: Precio predicho en ¬£
        """
        # Preprocesar
        X = self.preprocess(vehicle_data)
        
        # Predecir
        price = self.model.predict(X)[0]
        
        # Asegurar que el precio est√© en un rango razonable
        price = np.clip(price, 1000, 150000)
        
        return round(price, 2)
    
    def predict_with_confidence(self, vehicle_data, n_predictions=100):
        """
        Predecir con intervalo de confianza (usando bootstrap)
        
        Returns:
            dict: {'prediction': float, 'lower_bound': float, 'upper_bound': float}
        """
        predictions = []
        
        for _ in range(n_predictions):
            # A√±adir ruido peque√±o para simular incertidumbre
            noisy_data = vehicle_data.copy()
            noisy_data['mileage'] += np.random.randint(-500, 500)
            
            pred = self.predict(noisy_data)
            predictions.append(pred)
        
        return {
            'prediction': np.mean(predictions),
            'lower_bound': np.percentile(predictions, 5),
            'upper_bound': np.percentile(predictions, 95),
            'std_dev': np.std(predictions)
        }


# EJEMPLO DE USO:
if __name__ == '__main__':
    # Crear predictor con el modelo y scaler entrenados
    predictor = BMWPricePredictor(model=xgb_tuned, scaler_obj=scaler)
    
    # Datos de ejemplo
    vehicle = {
        'model': '3 Series',
        'year': 2018,
        'transmission': 'Automatic',
        'mileage': 35000,
        'fuelType': 'Diesel',
        'tax': 145,
        'mpg': 58.9,
        'engineSize': 2.0
    }
    
    # Predicci√≥n simple
    price = predictor.predict(vehicle)
    print(f"Precio predicho: ¬£{price:,.2f}")
    
    # Predicci√≥n con intervalo de confianza
    result = predictor.predict_with_confidence(vehicle)
    print(f"\nPredicci√≥n con confianza:")
    print(f"  Precio estimado: ¬£{result['prediction']:,.2f}")
    print(f"  Rango (90% confianza): ¬£{result['lower_bound']:,.2f} - ¬£{result['upper_bound']:,.2f}")
````

### Anexo C: Glosario de T√©rminos

**Machine Learning:**
- **Supervised Learning:** Aprendizaje con datos etiquetados
- **Regression:** Predicci√≥n de valores continuos
- **Feature Engineering:** Creaci√≥n de nuevas variables informativas
- **Overfitting:** Modelo memoriza training set, mal desempe√±o en test
- **Underfitting:** Modelo demasiado simple, no captura patrones
- **Cross-Validation:** T√©cnica para evaluar modelo con m√∫ltiples splits

**M√©tricas:**
- **R¬≤ Score:** Proporci√≥n de varianza explicada (0-1, mayor es mejor)
- **MAE:** Error absoluto medio (¬£, menor es mejor)
- **RMSE:** Ra√≠z del error cuadr√°tico medio (¬£, penaliza errores grandes)
- **MAPE:** Error porcentual absoluto medio (%, menor es mejor)

**Algoritmos:**
- **XGBoost:** Gradient Boosting optimizado, muy preciso
- **Random Forest:** Ensemble de √°rboles, robusto
- **SVR:** Support Vector Regression, kernel no lineal
- **Ridge:** Regresi√≥n lineal con regularizaci√≥n L2

**T√©cnicas:**
- **One-Hot Encoding:** Convertir categ√≥ricas a binarias
- **Target Encoding:** Codificar categ√≥ricas con estad√≠stica del target
- **StandardScaler:** Escalar a media=0, std=1
- **Grid Search:** B√∫squeda exhaustiva de hiperpar√°metros
